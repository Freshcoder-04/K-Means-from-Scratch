```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
from k_means import KMeans

import pyarrow.feather as feather
```

```python
data = pd.read_csv('data.csv')
```

Visualise the data

```python
plt.figure()
plt.scatter(data.iloc[:,0],data.iloc[:,1],s=10,c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

On visualising, we can say that there should be 3 clusters. Let's see if Elbow mehtod gives us the same result.

```python
kMeans = KMeans(100,3,1e-10)
```

```python
kMeans.ElbowMethod(data.iloc[:,:2])
```

We can clearly see that the elbow point is k=3

```python
KMeansNew = KMeans(max_iter=10000,k=3,conv_th=1e-10)
KMeansNew.fit(data.iloc[:,:2])
predicted_labels = KMeansNew.predict(data.iloc[:,:2])
```

```python
plt.figure()
plt.scatter(data.iloc[:,0], data.iloc[:,1],s=10,c='black')
plt.scatter(KMeansNew.centroids[:,0], KMeansNew.centroids[:,1],c='red')
plt.show()
```
